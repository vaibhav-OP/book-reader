# Importing packages

```python
import wave
from piper import PiperVoice
from playsound import playsound
```

# Loading voice pack with Piper

```python
voice = PiperVoice.load("./en_US-lessac-medium.onnx", use_cuda=True)
```

# Fetching text from a txt file and dividing into chunks (per line)

```python
text_string = ""

try:
    with open("text.txt", "r") as file:
        content = file.read()
        text_string = content
except:
    print("Something went wrong opening the file.")

chunks = text_string.splitlines()
```

# Feading chunks to piper and generating wav file and mapping to line

```python
file_map = {}

for i in range(0, len(chunks)):
    audio_file_name = f"audios/test{i}.wav"
    with wave.open(audio_file_name, "wb") as wav_file:
        voice.synthesize_wav(chunks[i], wav_file)

    file_map[i] = audio_file_name
```
